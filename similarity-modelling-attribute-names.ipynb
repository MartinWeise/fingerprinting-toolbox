```python
def find_perfect_match(list1, list2):   # list1 is referred to as the one with original names
    perfect_matches = []
    unmatched1 = list1
    unmatched2 = list2
    for element in list1:
        original_name = element
        #preprocess
        element = element.strip().lower()
        print(original_name)
        print(element)
        for element2 in unmatched2:
            original2 = element2
            element2 = element2.strip().lower().strip()
            if element == 'liability to equity':
                print(element)
                print(element2)
            if element == element2:
                perfect_matches.append(original_name)
                unmatched1.remove(original_name)
                unmatched2.remove(original2)
                print('______________________________')
                break
    return perfect_matches, (unmatched1, unmatched2)
```

```python
perfect_matches, (unmatched1, unmatched2) = find_perfect_match(list(names.name), list(data.columns))
```
